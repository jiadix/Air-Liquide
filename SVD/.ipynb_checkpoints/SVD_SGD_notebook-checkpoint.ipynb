```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from SVD import SVD_SGD
from SVD import SVD_SGD_python
from sklearn.model_selection import train_test_split
```

```python
# Loads movie data
df = pd.read_csv('../Data/ml-latest-small/ratings.csv')
max_U = np.max(df['userId'])
max_I = np.max(df['movieId'])
```

```python
# Separates into training, cross-validation and testing sets
p_test = 0.2
p_xvalid = 0.2
df_train, df_test = train_test_split(df, test_size=p_test)
df_train, df_xvalid = train_test_split(df_train,test_size=p_xvalid/(1-p_test))
```

```python
# Tests timing on SGD functions 
start = time.time()
SVD_SGD_python(df_train, n_factors=3, n_epochs=3);
end = time.time()
dt_python = end-start
print 'This function took %f seconds' % dt_python 

start = time.time()
SVD_SGD(df_train, n_factors=3, n_epochs=3);
end = time.time()
dt_cython = end-start
print 'This function took %f seconds' % dt_cython

print 'This is %f times faster!' % (dt_python/dt_cython)
```

```python
# users Id, movie Id, and rating cross-validation
users_xvalid = df_xvalid['userId']
movies_xvalid = df_xvalid['movieId']
r_xvalid = df_xvalid['rating']

# users Id, movie Id, and rating test set
users_test = df_test['userId']
movies_test = df_test['movieId']
r_test = df_test['rating']
```

```python
# Creates function to help with cross validation 
# Ouputs mean absolute error and model parameters
def SVDcv(lambda_u, lambda_i, lambda_p, lambda_t, n_fact,
         bu_cv=None, bi_cv=None, p_cv=None, t_cv=None):
    
    e_arr, mu, bu, bi, p, t = SVD_SGD(
        df_train,
        n_factors=n_fact, maxUserId=max_U, maxItemId=max_I,
        alpha=0.01, n_epochs=100, epsilon=0.1, e_tol=0.0005,
        lam_u=lambda_u, lam_i=lambda_i, lam_p=lambda_p, lam_t=lambda_t,
        bu_init=bu_cv, bi_init=bi_cv, p_init=p_cv, t_init=t_cv)
    
    r_preds = mu+bu[users_xvalid]+bi[movies_xvalid]
    r_preds += np.sum(p[users_xvalid]*t[movies_xvalid],axis=1)
    MAE = np.mean(np.abs(r_xvalid-r_preds))

    return (MAE, mu, bu, bi, p, t)
```

```python
# Sets up parameter sweep
param_arr = [[10.0**p_u, 10.0**p_i, 10.0**p_p, 10.0**p_t, 2**p_f] 
 for p_f in range(0,10)
 for p_u in range(-3,1) for p_i in range(-3,1) 
 for p_p in range(0,3) for p_t in range(0,3)]
df_params = pd.DataFrame(data=param_arr, 
             columns=['lambda_u', 'lambda_i',
                      'lambda_p', 'lambda_t', 'n_factors']);
```

```python
# Performs parameter sweep (this takes a while...)
# To speed up process, initializes initial solution at last
# solution. 
MAE_arr = []
n_fact_prev = 0
for idx, params in enumerate(param_arr):
    lam_u, lam_i, lam_p, lam_t, n_fact = params
    if idx == 0:
        MAE, mu, bu, bi, p, t = SVDcv(lam_u, lam_i, lam_p, lam_t, n_fact)
    elif n_fact_prev != n_fact:
        MAE, mu, bu, bi, p, t = SVDcv(lam_u, lam_i, lam_p, lam_t, n_fact,
                                     bu_cv=bu, bi_cv=bi)
    else:
        MAE, mu, bu, bi, p, t = SVDcv(lam_u, lam_i, lam_p, lam_t, n_fact,
                                     bu_cv=bu, bi_cv=bi, p_cv=p, t_cv=t)

    MAE_arr.append(MAE)
    n_fact_prev = n_fact
MAE_arr = np.array(MAE_arr)
df_params.loc[:,'MAE'] = pd.Series(MAE_arr)
df_params.to_csv('MAE_param_sweep.csv')
```

```python
# plots minimum MAE for each number of factors tried
plt.plot([2**p_f for p_f in range(0,10)],np.min(MAE_arr.reshape((10,144)),axis=1))
plt.xlabel('factors')
plt.ylabel('MAE')
plt.xscale('log')
plt.show()
idx = np.where(MAE_arr==np.min(MAE_arr))[0][0]
print param_arr[idx]
print MAE_arr[idx]
```

```python
# Finds mean absolute error of test set at best parameters
e_arr, mu, bu, bi, p, t = SVD_SGD(
    df_train,
    n_factors=128, maxUserId=max_U, maxItemId=max_I,
    alpha=0.01, n_epochs=100, epsilon=0.1, e_tol=0.0005,
    lam_u=0.01, lam_i=1.0, lam_p=10.0, lam_t=10.0)
    
r_preds = mu+bu[users_test]+bi[movies_test]
r_preds += np.sum(p[users_test]*t[movies_test],axis=1)
MAE = np.mean(np.abs(r_test-r_preds))

print MAE
```
